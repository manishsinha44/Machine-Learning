```python
# Load model directly
from transformers import AutoImageProcessor, AutoModelForImageClassification

processor = AutoImageProcessor.from_pretrained("microsoft/resnet-50")
model = AutoModelForImageClassification.from_pretrained("microsoft/resnet-50")
```

```python
import torch
```

```python
def classify_image(image):
  image = processor(image, return_tensors="pt")['pixel_values']
  logits = model(image).logits
  predicted_label=logits.argmax(-1).item()
  return model.config.id2label[predicted_label]
```

```python
iface = gr.Interface(fn=classify_image, inputs=gr.Image(type="pil"), outputs=gr.Label(), title = 'MANISH IMAGE CLASSIFIER USING RESNET-50', description="Image classifier built using Microsoft RESNET-50 model")
iface.launch()
```
