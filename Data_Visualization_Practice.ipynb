```python
investments_df = pd.read_csv('crypto_daily_prices.csv')
investments_df
```

```python
plt.plot('Date','BTC',data=investments_df)
```

```python
investments_df.plot(x='Date',y='BTC',label= 'Bitcoin Price', color='r',linewidth=3,figsize=(14,6))
plt.ylabel('Price[$]')
plt.xlabel('Date')
plt.title('Bitcoin Price')
plt.legend(loc= 'upper right')
plt.grid()
```

```python
investments_df.plot(x='Date',y=['BTC','ETH','ADA'],linewidth=2,figsize=(14,6))
plt.ylabel('Price[$]')
plt.xlabel('Date')
plt.title('Cryptop Prices')
plt.legend(loc= 'upper right')

plt.grid()
```

```python
investments_df.plot(x='Date',title='Crypto Prices',linewidth=2,subplots=True ,figsize=(15,20))
```

```python
daily_return_df = pd.read_csv('crypto_daily_returns.csv')
daily_return_df
```

```python
daily_return_df.plot.scatter('BTC','ETH',grid=True,figsize=(12,8),color = 'hotpink',alpha=0.3)
```

```python
daily_return_df.plot.scatter('BTC','ADA',grid=True,figsize=(12,8),color = 'coral',alpha=0.4)
```

```python
mu = round(daily_return_df['BTC'].mean(),2)
sigma = round(daily_return_df['BTC'].std(),2)
num_bins = 40
daily_return_df[['BTC','ETH']].plot.hist(bins=60,alpha=0.5,figsize=(12,8),grid=True,subplots=True)
plt.xlabel('Returns')
plt.title('Histogram of Crypto Returns')
```
