```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
```

```python
audio_path = "example.mp3"
```

```python
y,sr = librosa.load(audio_path,sr=None)
```

```python
type(y)
```

```python
y.shape
```

```python
sr   # in Hz
```

48000 Hz = 48 kHz ; Hz = 1/second. This essentially means we are collecting 48000 samples of information per second.
Since y has total 630031 samples, we have around 630031/48000 = 13 (approx) seconds of audio.

```python
plt.figure(figsize=(14,5),dpi = 150)
plt.plot(y)
plt.xlabel("Time - Samples")
plt.ylabel("Amplitude")
plt.show()
```

```python
from IPython.display import Audio
Audio(data=y, rate=sr)
```

```python
# DFT - Discrete Fourier Transform .This is the way we convert time to frequncy
window = np.hanning(len(y))
windowed_output = y * window
dft = np.fft.rfft(windowed_output)
```

```python
amplitude = np.abs(dft)
```

```python
amplitude_db = librosa.amplitude_to_db(amplitude, ref=np.max)
```

```python
freq = librosa.fft_frequencies(sr=sr, n_fft=len(y))
```

```python
plt.figure(figsize=(14,5),dpi = 150)
plt.plot(freq, amplitude_db)
plt.xlabel("Frequency Hz")
plt.ylabel("Amplitude in db")
plt.xscale('log')
plt.show()
```
