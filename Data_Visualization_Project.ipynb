```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
price_df = pd.read_csv('stocks_daily_prices.csv')
returns_df = pd.read_csv('stocks_daily_returns.csv')
```

```python
price_df.head()
```

```python
returns_df.head()
```

```python
price_df.plot(x="Date",legend=True,figsize=(25,20),subplots=True,layout=(4,4))
```

```python
price_df.plot(x="Date",legend=True,figsize=(15,20),subplots=True)
```

```python
price_df.plot(x="Date",legend=True,figsize=(15,10),grid=True)
```

```python
returns_df.plot.scatter(x="AAPL",y="JPM",color="teal",alpha=0.5,figsize=(13,10))
```

```python
plt.figure(figsize=(13,10))
sns.scatterplot(data=returns_df,x="AAPL",y="JPM",color="teal",alpha=0.5)
```

```python
alloc_df = pd.DataFrame(data=[70,10,10,10],index=["AAPL","JPM","MSFT","UAPL"],columns=["Allocation"])
alloc_df
```

```python
alloc_df.plot.pie(y="Allocation",explode=[0.2,0,0,0],figsize=(10,10),autopct="%1.1f%%",shadow=True,colormap="YlGnBu")
```

```python
returns_df[["UAL","PG"]].plot.hist(bins=40,figsize=(15,10),alpha=0.5,grid=True)
```

```python
sns.heatmap(returns_df[['AAPL','JPM','UAL','PG']].corr(),annot=True,cmap="YlGnBu")
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111,projection="3d")
x = returns_df['AAPL']
y = returns_df['JPM']
z = returns_df['UAL']
ax.scatter(x,y,z,c='r',marker='o',s=50)
ax.set_xlabel("AAPL")
ax.set_ylabel("JPM")
ax.set_zlabel("UAL")
```
